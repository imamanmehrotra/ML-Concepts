# Min Max Normalization

### It tries to scale down the value of features between 0 to 1

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
url='https://raw.githubusercontent.com/krishnaik06/Machine-Learning-in-90-days/master/Section%201-%20Python%20Crash%20Course/wine.csv'
ndf=pd.read_csv(url, usecols=[1,2,3])
ndf.columns=['Class','Alcohol','Malic']
scaling=MinMaxScaler()
norm=scaling.fit_transform(ndf[['Alcohol','Malic']])
n=pd.DataFrame(norm)
n.columns=['Scaled Down Value for Alcohol','Scaled Down Value for Malic']
```

# Standardization, also known as Z-Score Normalization

### Using Z-Score Normalization, all the features here will exhibit the feature of Standard normal distribution where Mean =0 and Standard_Deviation=1

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaling=StandardScaler()
```

```python
stdzn=pd.DataFrame(scaling.fit_transform(ndf[['Alcohol','Malic']]))
stdzn.columns=['St Alcohol','St Malic']
stdzn
```
