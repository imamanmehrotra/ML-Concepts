```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('titanic.csv', usecols=['Survived','Cabin'])
df
```

```python
df['Cabin'].fillna('Missing', inplace=True)
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
```

```python
df
```

```python
pr_survived=df.groupby(['Cabin'])['Survived'].mean().sort_values(ascending=False)
pr_survived
```

```python
pr_died=1-pr_survived
pr_died
```

```python
pr_ratio=pr_survived/pr_died
pr_dict=pr_ratio.to_dict()
```

```python
df['Encoded_probability_ratio']=df['Cabin'].map(pr_dict)
```
