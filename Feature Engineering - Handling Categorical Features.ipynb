```python
df=pd.read_csv('mb.csv', usecols=['X0','X1','X2','X3','X4','X5','X6'])
df.head()
```

```python
for i in df.columns:
    print(df[i].value_counts().head(10),'\n')
```

```python
lst_10=df['X0'].value_counts().sort_values(ascending=False).head(10).index.tolist()
lst_10
```

```python
import numpy as np
for cat in lst_10:
    df[cat]=np.where(df['X0']==cat,1,0)
```
