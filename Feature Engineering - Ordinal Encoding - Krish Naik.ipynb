```python
url='https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
new=pd.read_csv(url, header=None, index_col=None)
new.head(10)
```

```python
ts
```

```python
for ftr in ts.columns:
    print(ftr,"has :",len(ts[ftr].unique()),' different types of labels')
```
